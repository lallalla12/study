```python
# 지도에 특정 지역만 표시
import pandas as pd
import folium

mymap=folium.Map(location=[33.371296, 126.560056],zoom_start=11)
df=pd.read_excel('전국도시공원정보표준데이터.xls')
df=df.dropna(subset=['위도','경도','소재지지번주소'])
df=df.query("소재지지번주소.str.contains('제주')",engine='python')

for i in range(len(df)):
    folium.Marker([float(df['위도'][i:i+1]),
                   float(df['경도'][i:i+1])],
                  popup=df['공원명'][i:i+1]).add_to(mymap)
mymap
```

```python
# 공원뒤에 데이터타입도 같이 출력됨
for i in range(len(df)):
    print(float(df['위도'][i:i+1]), float(df['경도'][i:i+1]), df['공원명'][i:i+1])
```

```python
# 필요한 글자만 출력되도록 iloc를 사용
for i in range(len(df)):
    folium.Marker([float(df.iloc[i,5]),float(df.iloc[i,6])],popup=df.iloc[i,1]).add_to(mymap)
mymap
```

```python
import pandas as pd
import folium

# page.193 mission
# 1. input()함수를 이용해사용자가 원하는 지역을 입력하여 지도로 표시

df = pd.read_excel('전국도시공원정보표준데이터_신버전.xlsx')
df = df.dropna(subset=['위도', '경도', '소재지지번주소'])

# 사용자 지역 입력
name = input('지역명을 입력하세요: ').strip()
df_filtered = df[df["소재지지번주소"].astype(str).str.contains(name, na=False)]

# 지도 중심 이동
if not df_filtered.empty:
    center_lat = float(df_filtered['위도'].iloc[0])
    center_lon = float(df_filtered['경도'].iloc[0])
    mymap = folium.Map(location=[center_lat, center_lon], zoom_start=12)

    for i in range(len(df_filtered)):
        lat = float(df_filtered["위도"].iloc[i])
        lon = float(df_filtered["경도"].iloc[i])
        park = df_filtered["공원명"].iloc[i]
        folium.Marker([lat, lon], popup=park).add_to(mymap)

else:
    print("해당 지역의 공원이 존재하지 않습니다.")
    mymap = folium.Map(location=[33.371296, 126.560056], zoom_start=11)

mymap
```
