```python
# page. 151 mission
# numpy를 이용해서 생성한 값에 대한 선점도 그래프 그리기
# numpy 사용하기
import numpy as np
# 차트를 그리기 위해 matplotlib 사용하기
import matplotlib.pyplot as plt

# x축 필요한 값 : 1~10까지 (arange)
x = np.arange(1,11) 

# y축 필요한 값 : 1~30까지 임의의 정수 10개
y = np.random.randint(1, 30, 10)
print(x)
print(y)

plt.scatter(x, y, c='r', marker='s')
plt.show()
```

```python
# page.159 mission
# 학생별 국어 점수와 수학 점수를 선 그래프로 그리기
# 차트를 그리기 위해 matplotlib 사용하기
import matplotlib.pyplot as plt

# 학생별 국어 점수와 수학 점수를 저장한 data1.xlsx파일을 파이썬으로 가져오기 위한 pandas 사용하기
import pandas as pd

# 한글 깨짐 방지
plt.rc('font', family='Gulim')
df = pd.read_excel('data1.xlsx')

# x축에 필요한 값
x = df['name'] 
# y축에 필요한 첫번재 값
y1 = df['kor']  
# y축에 필요한 두번재 값
y2 = df['math']  

plt.plot(x, y1, 'g--', marker='o', label='국어점수')
plt.plot(x, y2, 'r-', marker='^', label='수학점수')
plt.title('성적 그래프')
plt.xlabel('이름')
plt.ylabel('점수')
plt.legend()
```
