```python
# 산점도 그래프 : scatterplot()
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips_data)
plt.title('Total bill & tips')
plt.show()
```

```python
# 범주별 데이터 개수를 표시하는 막대 그래프 : counterplot()
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')
sns.countplot(x='day', data=tips_data)
plt.show()
```

```python
# 범주별 항목값 평균을 표시하는 막대 그래프 : barplot()
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')
sns.barplot(x='day', y='tip', data=tips_data)
plt.show()
```

```python
# 히스토그램 : histplot()
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.histplot(x='tip', data=tips_data, kde=True)
plt.show()
```

```python
# 숫자값을 갖는 칼럼들의 상관관계를 보여주는 그래프 : pairplot()
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.pairplot(tips_data, hue='time')
plt.show()
```

```python
# 산점도와 추세선(회귀선)을 동시에 보여주는 그래프 : regplot()
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.regplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
# 집계한 값에 따라서 색깔을 표시하는 그래프(히트맵) : heatmap()
import matplotlib.pyplot as plt
import seaborn as sns

tips_data = sns.load_dataset('tips')
df = tips_data.corr(numeric_only=True)

sns.heatmap(data=df, annot=True, fmt='.2f', cbar=False)
plt.show()
```

```python
# page.169 mission2
# 1. load_dataset()을 이용해서 penguins 데이터 가져오기
import matplotlib.pyplot as plt
import seaborn as sns

penguins_data = sns.load_dataset('penguins')
sns.barplot(x='species', y='body_mass_g', data=penguins_data, hue='sex')
plt.show()
```

```python
# 연도별 총 인구수, 출생아 수를 하나의 차트에 그리기
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel("총인구수출생아수.xlsx")

fig,ax1=plt.subplots()
plt.rc('font', family='Gulim')

plt.title('연도별 총인구수와 출생아수   단위:만명', pad=20)
ax1.set_xlabel('연도')
ax1.set_ylabel('총인구수')
ax1.plot(df['연도'], df['총인구수']/10000,linestyle='--')

# 같은 x축을 공유하는 두 번째 축
ax2=ax1.twinx()
ax2.set_ylabel('출생아수',color='blue')
ax2.plot(df['연도'], df['출생아수']/10000,color='blue')
ax2.tick_params(axis='y',labelcolor='blue')

fig.tight_layout()
plt.show()
```
